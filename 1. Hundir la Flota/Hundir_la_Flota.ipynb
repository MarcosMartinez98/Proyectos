```python

import numpy as np
import random
import time
```

```python
class Tablero:

    def __init__(self, id_jugador, tamaño_tablero, esloras, dificultad):
        self.id_jugador = id_jugador
        self.tamaño_tablero = tamaño_tablero
        self.tablero_jugador = np.full((tamaño_tablero,tamaño_tablero), '-')
        self.tablero_maquina = np.full((tamaño_tablero,tamaño_tablero), '-')
        self.esloras = esloras # Se introducen con una lista: [Es.4,Es.3,Es.2,Es1]
        self.dificultad = dificultad
        self.tablero_interno_jugador = np.full((10,10), '-') 
        self.barcos_jugador = []
        self.barcos_maquina = []
        
    def tablero_bonito(self, tabla): # Imprime el tablero en pantalla de forma legible, con separaciones por filas y columnas.
        count = 0
        for i in range(self.tamaño_tablero):
            for j in range(self.tamaño_tablero):
                count += 1
                print(tabla[i,j], end = ' ')
                if count % self.tamaño_tablero == 0:
                    print()

    def limpiar_tablero(self): # Sustituye los asteriscos ('*') por guiones ('-') para "limpiar" el tablero visualmente.
        self.tablero_jugador[self.tablero_jugador == '*'] = '-'
        self.tablero_maquina[self.tablero_maquina == '*'] = '-'

    def crear_barcos(self, tablero, barcos): 
        
        '''Selecciona una orientación aleatoria: Este (E), Oeste (O), Norte (N), Sur (S).

Verifica si el barco cabe en el tablero y no se solapa con otros, coloca el barco ('O') y lo rodea de asteriscos ('*') para marcar espacio ocupado.

Almacena las posiciones del barco y su estado (vidas, tocados) en la lista correspondiente.'''


        count4 = 0 
        barcos_eslora4 = []
        while count4 < self.esloras[0]:
            
            orientacion4 = random.choice(['E','N','O','S'])
            
            # Barco tamaño 4
            
            if orientacion4 == 'E':
                x = int(random.choice(np.arange(self.tamaño_tablero)))
                y = int(random.choice(np.arange(self.tamaño_tablero-3)))
                for posicion in [(x,y), (x,y+1), (x,y+2), (x,y+3)]: 
                    tablero[posicion] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1), (x-1,y+2), (x-1,y+3), (x-1,y+4),(x,y-1),(x,y+4),\
                                    (x+1,y-1), (x+1,y), (x+1,y+1), (x+1,y+2), (x+1,y+3), (x+1,y+4)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*'
                count4 += 1
                barcos_eslora4.append({'vidas': [(x,y), (x,y+1), (x,y+2), (x,y+3)], 'tocados': []})

            elif orientacion4 == 'O':
                x = int(random.choice(np.arange(self.tamaño_tablero)))
                y = int(random.choice(np.arange(3,self.tamaño_tablero)))
                for posicion in [(x,y), (x,y-1), (x,y-2), (x,y-3)]: 
                    tablero[posicion] = 'O'
                for fila, columna in [(x-1,y+1), (x-1,y), (x-1,y-1), (x-1,y-2), (x-1,y-3), (x-1,y-4),(x,y+1),(x,y-4),\
                                    (x+1,y+1), (x+1,y), (x+1,y-1), (x+1,y-2), (x+1,y-3), (x+1,y-4)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count4 += 1
                barcos_eslora4.append({'vidas':[(x,y), (x,y-1), (x,y-2), (x,y-3)], 'tocados': []})

            elif orientacion4 == 'N':
                x = int(random.choice(np.arange(3,self.tamaño_tablero)))
                y = int(random.choice(np.arange(self.tamaño_tablero)))
                for posicion in [(x,y), (x-1,y), (x-2,y), (x-3,y)]: 
                    tablero[posicion] = 'O'
                for fila, columna in [(x+1,y-1), (x+1,y), (x+1,y+1),(x,y-1),(x,y+1),(x-1,y-1),(x-1,y+1),\
                                    (x-2,y-1),(x-2,y+1),(x-3,y-1),(x-3,y+1),(x-4,y-1), (x-4,y), (x-4,y+1)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count4 += 1
                barcos_eslora4.append({'vidas':[(x,y), (x-1,y), (x-2,y), (x-3,y)], 'tocados': []})
                
            elif orientacion4 == 'S':
                x = int(random.choice(np.arange(self.tamaño_tablero - 3)))
                y = int(random.choice(np.arange(self.tamaño_tablero)))
                for posicion in [(x,y), (x+1,y), (x+2,y), (x+3,y)]: 
                    tablero[posicion] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1),(x,y-1),(x,y+1),(x+1,y-1),(x+1,y+1),\
                                    (x+2,y-1),(x+2,y+1),(x+3,y-1),(x+3,y+1),(x+4,y-1), (x+4,y), (x+4,y+1)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count4 += 1
                barcos_eslora4.append({'vidas':[(x,y), (x+1,y), (x+2,y), (x+3,y)], 'tocados': []})
        barcos.append(barcos_eslora4)
        
        # Barcos de tamaño 3
        
        count3 = 0
        barcos_eslora3 = []
        while count3 < self.esloras[1]:
            
            orientacion3 = random.choice(['E','N','O','S'])
            if orientacion3 == 'E':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero-2)))
                    posiciones = [(x, y), (x, y+1), (x, y+2)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones):
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1), (x-1,y+2), (x-1,y+3),\
                                    (x,y-1),(x,y+3),(x+1,y-1), (x+1,y), (x+1,y+1), (x+1,y+2), (x+1,y+3)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count3 += 1
                barcos_eslora3.append({'vidas':[(x, y), (x, y+1), (x, y+2)], 'tocados':[]})
            
        
                
            elif orientacion3 == 'O':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(2,self.tamaño_tablero)))
                    posiciones = [(x, y), (x, y-1), (x, y-2)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y+1), (x-1,y), (x-1,y-1), (x-1,y-2), (x-1,y-3),\
                                    (x,y+1),(x,y-3),(x+1,y+1), (x+1,y), (x+1,y-1), (x+1,y-2), (x+1,y-3)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count3 += 1
                barcos_eslora3.append({'vidas':[(x, y), (x, y-1), (x, y-2)], 'tocados':[]})
                
            elif orientacion3 == 'N':
                while True:
                    x = int(random.choice(np.arange(2,self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero)))
                    posiciones = [(x, y), (x-1, y), (x-2, y)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x+1,y-1), (x+1,y), (x+1,y+1),(x,y-1),(x,y+1),\
                                    (x-1,y-1),(x-1,y+1),(x-2,y-1),(x-2,y+1),(x-3,y-1), (x-3,y), (x-3,y+1)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count3 += 1
                barcos_eslora3.append({'vidas':[(x, y), (x-1, y), (x-2, y)], 'tocados':[]})
                
            elif orientacion3 == 'S':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero-2))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero)))
                    posiciones = [(x, y), (x+1, y), (x+2, y)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones):
                        break 
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1),(x,y-1),(x,y+1),
                                    (x+1,y-1),(x+1,y+1),(x+2,y-1),(x+2,y+1),(x+3,y-1), (x+3,y), (x+3,y+1)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count3 += 1
                barcos_eslora3.append({'vidas':[(x, y), (x+1, y), (x+2, y)], 'tocados':[]})
        barcos.append(barcos_eslora3)

        # Barcos de tamaño 2
        
        count2 = 0
        barcos_eslora2 = []
        while count2 < self.esloras[2]:
            
            orientacion2 = random.choice(['E','N','O','S'])
            
            if orientacion2 == 'E':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero-1)))
                    posiciones = [(x, y), (x, y+1)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1), (x-1,y+2),\
                                    (x,y-1),(x,y+2),(x+1,y-1), (x+1,y), (x+1,y+1), (x+1,y+2)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count2 += 1
                barcos_eslora2.append({'vidas':[(x, y), (x, y+1)], 'tocados':[]})

            elif orientacion2 == 'O':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(1,self.tamaño_tablero)))
                    posiciones = [(x, y), (x, y-1)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y+1), (x-1,y), (x-1,y-1), (x-1,y-2),(x,y+1),(x,y-2),
                                    (x+1,y+1), (x+1,y), (x+1,y-1), (x+1,y-2)]: #Asteriscos
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' 
                count2 += 1
                barcos_eslora2.append({'vidas':[(x, y), (x, y-1)], 'tocados':[]})

            elif orientacion2 == 'N':
                while True:
                    x = int(random.choice(np.arange(1,self.tamaño_tablero))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero)))
                    posiciones = [(x, y), (x-1, y)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x+1,y-1), (x+1,y), (x+1,y+1),(x,y-1),(x,y+1),\
                                    (x-1,y-1),(x-1,y+1),(x-2,y-1),(x-2,y), (x-2,y+1)]:
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' #asteriscos
                count2 += 1
                barcos_eslora2.append({'vidas':[(x, y), (x-1, y)], 'tocados':[]})

            elif orientacion2 == 'S':
                while True:
                    x = int(random.choice(np.arange(self.tamaño_tablero-1))) 
                    y = int(random.choice(np.arange(self.tamaño_tablero)))
                    posiciones = [(x, y), (x+1, y)]
                    if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                        break  
                for pos in posiciones:
                    tablero[pos] = 'O'
                for fila, columna in [(x-1,y-1), (x-1,y), (x-1,y+1),(x,y-1),(x,y+1),\
                                    (x+1,y-1),(x+1,y+1),(x+2,y-1),(x+2,y),(x+2,y+1)]:
                    if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                        tablero[fila, columna] = '*' #asteriscos
                count2 += 1
                barcos_eslora2.append({'vidas':[(x, y), (x+1, y)], 'tocados':[]})
        barcos.append(barcos_eslora2)
        
        # Barcos de tamaño 1
        
        count1 = 0 # Aquí no hay orientacion y siempre van a caber dentro del tablero.
        barcos_eslora1 = []
        while count1 < self.esloras[3]:
            
            while True:
                x = int(random.choice(np.arange(self.tamaño_tablero))) 
                y = int(random.choice(np.arange(self.tamaño_tablero)))
                posiciones = [(x, y)]
                if all(tablero[pos] != 'O' and tablero[pos] != '*' for pos in posiciones): 
                    break  
            for pos in posiciones:
                tablero[pos] = 'O'
            for fila, columna in [(x-1,y-1),(x,y-1),(x+1,y-1), (x-1,y),(x+1,y), (x-1,y+1), (x,y+1), (x+1,y+1)]: #Asteriscos
                if 0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero:
                    tablero[fila, columna] = '*' 
            count1 += 1
            barcos_eslora1.append({'vidas':[(x, y)], 'tocados':[]})
        barcos.append(barcos_eslora1)

    def disparo_jugador(self):
        
        '''Solicita coordenadas al jugador para disparar.

Verifica que el formato del input sea el adecuado y que la posición no se haya usado antes.

Marca impacto si hay barco enemigo ('O') con 'X', o con 'A' si impacta en agua.

Actualiza el estado del barco (quita de vidas, añade a tocados) e informa si un barco ha sido hundido (cuando vidas queda vacío).'''

        while True:
            print()
            print(f"¡Es tu turno, {self.id_jugador}!")
            print()
            print("TABLERO ENEMIGO")
            self.tablero_bonito(self.tablero_interno_jugador)
            coordenadas_seleccionadas = input(f"Introduce coordenadas (fila,columna) entre 0 y {self.tamaño_tablero}: ").strip()
            
            # VALIDAR QUE HAYA EXACTAMENTE UNA COMA Y DOS NÚMEROS
            
            if ',' not in coordenadas_seleccionadas:
                print("Introduce un formato válido (ej: 3,5).")
                continue
            coordenadas = coordenadas_seleccionadas.split(',')
            if len(coordenadas) != 2:
                print("Debes ingresar exactamente dos números separados por una coma "," (ej: 3,5).")
                continue
            
            coordenada_1 = coordenadas[0]
            coordenada_2 = coordenadas[1]

            
            fila = int(coordenada_1)
            columna = int(coordenada_2)

            # VALIDAR QUE LAS COORDENADAS EXISTAN EN EL TABLERO 
            
            if not (0 <= fila < self.tamaño_tablero and 0 <= columna < self.tamaño_tablero):
                print("Coordenadas fuera de rango. Introduce valores entre 0 y 9.")
            
            punto_impacto = self.tablero_maquina[fila, columna]

            # COMPRUEBA DÓNDE HA IMPACTADO, Y LO QUE SUCEDE EN CADA SITUACIÓN

            if punto_impacto == 'X' or punto_impacto == 'A':
                print("Ya disparaste ahí. Intenta otra vez.")
            
            elif punto_impacto == 'O':
                print("¡TOCADO! ¡Has impactado en un barco enemigo!")
                self.tablero_maquina[fila, columna] = 'X'
                self.tablero_interno_jugador[fila,columna] = 'X'
                for i in self.barcos_maquina:
                    for j in i:
                        if (fila, columna) in j['vidas']:
                            j['tocados'].append(j['vidas'].pop(j['vidas'].index((fila,columna))))
                            for i in self.barcos_maquina:
                                for j in i:
                                    if len(j['vidas']) == 0:
                                        print(f'Barco enemigo de eslora {len(j["tocados"])} hundido')
                                    
                if list(self.tablero_maquina[self.tablero_maquina == 'O']).count('O') == 0:
                    break
                        
            else:
                print()
                print("¡Tu disparo ha dado en el agua!")
                print()
                print("·-·-·"*10)
                print()
                self.tablero_maquina[fila, columna] = 'A'
                self.tablero_interno_jugador[fila,columna] = 'A'
                break 
    
    def disparo_maquina(self):
            

        # EMPIEZA EL TURNO DE LA MÁQUINA en función de la dificultado
        
        if self.dificultad == 'Normal':
            
            self.ultimo_impacto_maquina = None

            print("\nTurno de la máquina\n")
            time.sleep(2)

            while True:
                
                # Si la máquina impacta su último tiro, el siguiente lo intenta en adyacentes a esa casilla. (El primer tiro siempre es aleatorio).
                
                if self.ultimo_impacto_maquina:
                    fila, columna = self.ultimo_impacto_maquina
                    adyacentes = [(fila-1, columna), (fila+1, columna), (fila, columna-1), (fila, columna+1)] # Todas las aydacentes
                    random.shuffle(adyacentes)  # Para que no siga siempre el mismo patrón

                    # Sólo se considera coordenadas dentro del tablero y que no se repitan
                    
                    for f, c in adyacentes:
                        if 0 <= f < self.tamaño_tablero and 0 <= c < self.tamaño_tablero:
                            if self.tablero_jugador[f, c] not in ['X', 'A']:  # No repetir disparos
                                break
                    else:
                        # Si no hay adyacentes válidos, hacer disparo aleatorio
                        self.ultimo_impacto_maquina = None
                        continue
                else:
                    # Primer disparo, aleatorio. Repite el disparo si cae en "X" o "A" hasta encontrar uno válido
                    f, c = np.random.randint(self.tamaño_tablero), np.random.randint(self.tamaño_tablero)
                    while self.tablero_jugador[f, c] in ['X', 'A']:
                        f, c = np.random.randint(self.tamaño_tablero), np.random.randint(self.tamaño_tablero)

                print(f"\nLa máquina dispara a ({f}, {c})\n")
                time.sleep(2) 

                #Si impacta en barco ("O"), guarda las coordednadas para calcular las adyacentes en su siguiente disparo
                if self.tablero_jugador[f, c] == 'O':
                    print("\n¡La máquina ha tocado uno de tus barcos!\n")
                    self.tablero_jugador[f, c] = 'X'
                    self.ultimo_impacto_maquina = (f, c)
                    
                    for i in self.barcos_jugador:
                        for j in i:
                            if (f, c) in j['vidas']:
                                j['tocados'].append(j['vidas'].pop(j['vidas'].index((f, c))))
                                if len(j['vidas']) == 0:
                                    print(f"\nLa máquina ha hundido tu barco de eslora {len(j['tocados'])}\n")
                                    self.ultimo_impacto_maquina = None  # Reinicia si el barco fue hundido
                                    
                    # Verifica si ya no quedan barcos del jugador                
                    if list(self.tablero_jugador[self.tablero_jugador == 'O']).count('O') == 0:
                        print("\n¡La máquina ha ganado!\n")
                        break
                # Si el disparo cae en agua ("A"), marca la casilla.
                else:
                    print("\nLa máquina ha fallado.\n")
                    self.tablero_jugador[f, c] = 'A'
                    self.ultimo_impacto_maquina = None #Reinicia el "None", por lo que no recordará el último impacto (Posible upgrade)
                    print('\nTU TABLERO:\n')
                    self.limpiar_tablero()
                    self.tablero_bonito(self.tablero_jugador)
                    break
        
        elif self.dificultad == 'Dificil':
            while True:
                print("\nTurno de la máquina\n")
                time.sleep(2)

                fila = np.random.randint(self.tamaño_tablero)
                columna = np.random.randint(self.tamaño_tablero)

                # Coordenadas del cañonazo, son la coordenada generada aleatoriamente y las casillas adyacentes
                coordenadas_disparo = [
                    (fila, columna), 
                    (fila, columna-1), 
                    (fila-1, columna-1), 
                    (fila-1, columna),
                    (fila-1, columna+1), 
                    (fila, columna+1),
                    (fila+1, columna-1), 
                    (fila+1, columna), 
                    (fila+1, columna+1)
                ]

                # Filtrar coordenadas válidas. Habra algunas adyacentes que se salgan fuera, pues las filtramos
                coordenadas_validas = [
                    (x, y) for x, y in coordenadas_disparo
                    if 0 <= x < self.tamaño_tablero and 0 <= y < self.tamaño_tablero
                ]

                tocado = False  # Controla si ha habido impacto

                for x, y in coordenadas_validas:
                    casilla = self.tablero_jugador[x, y]

                    if casilla == 'O':  # Barco impactado
                        print("\n¡Uno de tus barcos ha sido tocado!\n")
                        self.tablero_jugador[x, y] = 'X'
                        tocado = True

                        # Actualizar vidas y tocados
                        for barco in self.barcos_jugador:
                            for j in barco:
                                if (x, y) in j['vidas']:
                                    j['tocados'].append(j['vidas'].pop(j['vidas'].index((x, y))))
                                    if len(j['vidas']) == 0:
                                        print(f"\nUn barco de eslora {len(j['tocados'])} hundido.\n")

                    elif casilla == '-':  
                        self.tablero_jugador[x, y] = 'A'

                    # Si ya era 'X' o 'A', no se hace nada (ya disparado)

                self.limpiar_tablero()
                print('\nTU TABLERO:\n')
                self.tablero_bonito(self.tablero_jugador)

                # Fin del juego si no quedan barcos
                if list(self.tablero_jugador[self.tablero_jugador == 'O']).count('O') == 0:
                    break

                # Esto es la magia, si ha habido tocado, se ha cambiado a True y ahora not True = False y no hay break y la maquina sigue
                if not tocado:
                    break
```

```python
def jugar(nombre, tamaño_tablero, num_barcos, dificultad):

    # LLAMA A LA CLASE
    tableri = Tablero(nombre, tamaño_tablero, num_barcos, dificultad)  

    # CREAMOS LOS BARCOS DEL JUGADOR
    tableri.crear_barcos(tableri.tablero_jugador, tableri.barcos_jugador)

    # CREAMOS LOS BARCOS DE LA MAQUINA
    tableri.crear_barcos(tableri.tablero_maquina, tableri.barcos_maquina)

    while True: #Inicia los turnos, y comprueba tras cada uno de ellos si queda algún barco en su mapa correspondiente, si no es asi, finaliza el juego.

        tableri.disparo_jugador()

        if list(tableri.tablero_maquina[tableri.tablero_maquina == 'O']).count('O') == 0: 
            print('HAS GANADO')
            break

        tableri.disparo_maquina()

        if list(tableri.tablero_jugador[tableri.tablero_jugador == 'O']).count('O') == 0:
            print('HAS PERDIDO')
            break
```
